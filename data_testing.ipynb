```python
import fastf1 as ff1
import pandas as pd 
from cache_path import cache_path

ff1.Cache.enable_cache(cache_path)

rounds = range(1,5)
all_results = []

for round in rounds:
    session = ff1.get_session(2025, round, 'Q')
    session.load()

    result = session.results[['Position', 'Time']]
    result.index = result.index.astype(int)
    # Replace 'R', 'D', 'W' with 20
    # result = pd.to_numeric(result, errors="coerce").fillna(20)

    race_df = pd.DataFrame({
        "DriverNumber": result.index,
        "RoundNumber": round,
        "Position": result['Position'],
        'Time': result['Time']
    })

    race_df = race_df.sort_values('DriverNumber')

    all_results.append(race_df)

results = pd.concat(all_results, ignore_index=True)
results
```

```python
session = ff1.get_session(2025, 1, 'Q')
session.load()
session.results
```

```python
rows = []
for driver in session.drivers:
    lap = session.laps.pick_drivers(int(driver)).pick_fastest()
    if lap is None:
        continue 
    else:
        time = lap['LapTime']
        rows.append({
            'DriverNumber': int(driver),
            'RoundNumber': session.event['RoundNumber'],
            'FastestLapTime': time
        })
qual_times = pd.DataFrame(rows)
qual_times
```
